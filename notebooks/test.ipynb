```python
print("Any rows where all ydstogo buckets are 0?",
      ((df_feat["short_ydstogo"] == 0) &
       (df_feat["medium_ydstogo"] == 0) &
       (df_feat["long_ydstogo"] == 0)).any())

print("Value counts for is_red_zone:")
print(df_feat["is_red_zone"].value_counts(dropna=False))
```

```python
from nfl_run_pass.data_loading import load_and_prepare_run_pass_data
from nfl_run_pass.preprocessing import prepare_train_test_data
from nfl_run_pass.models import train_log_reg_model

df_model = load_and_prepare_run_pass_data()
X_train, X_test, y_train, y_test, scaler, feature_cols = prepare_train_test_data(df_model)

model, info = train_log_reg_model(X_train, y_train)

print("Used grid search:", info["used_grid_search"])
print("Best params:", info["best_params"])
print("Best CV score:", info["best_score"])

# sanity check: training predictions
from sklearn.metrics import accuracy_score

y_train_pred = model.predict(X_train)
print("Train accuracy:", accuracy_score(y_train, y_train_pred))
```
